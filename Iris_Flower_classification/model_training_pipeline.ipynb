```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pickle
```

```python
Labels={
    "Iris-setosa":0,
    "Iris-versicolor":1,
    "Iris-virginica":2
}
    
```

```python
data=pd.read_csv("IRIS.csv")
data.sample(4)
```

```python
features=data.drop('species',axis=1).to_numpy()
target=data['species'].replace(Labels).to_numpy()
```

```python
#scaling
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_features,target,test_size=0.2,stratify=target,shuffle=True,random_state=42)
```

```python
# Training
logistic_reg=LogisticRegression(class_weight='balanced')
logistic_reg.fit(X_train,y_train)
```

```python
# Evaluation
pred=logistic_reg.predict(X_test)
print(classification_report(y_test,pred))
```

```python
# model saving
#1.scaler
pickle.dump(scaler,open("standard_scaler.pkl",'wb'))
#2.model
pickle.dump(logistic_reg,open("model.pkl",'wb'))
```
